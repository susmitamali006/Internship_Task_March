```python
# Import important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Standardise column names 
df.columns = (
    df.columns 
    .str.lower()
    .str.strip()
    .str.replace(" ","_")
)
df.columns
```

```python
# Null value Count
print(df.isnull().sum())
```

```python
# null value handling for column having data type numeric
df['data_value'] = df['data_value'].fillna(df['data_value'].median())

#confidence_limit_low with its median
low_median = df['confidence_limit_low'].median()
df['confidence_limit_low'] = df['confidence_limit_low'].fillna(low_median)

#confidence_limit_High with its median
high_median = df['confidence_limit_high'].median()
df['confidence_limit_high'] = df['confidence_limit_high'].fillna(high_median)
```

```python
# null value handling for column having data type object
df['response'] = df['response'].fillna("Unknown")

# Drop the footnote columns as they are not needed for calculation
df = df.drop(columns=['data_value_footnote_symbol', 'data_value_footnote'])
```

```python
# After null value handling 
print("After Null Value Handling Count:\n",df.isnull().sum()) 
```

```python
# Count the Duplicates 
print(df.duplicated().sum())
```

```python
# Get a summary of numerical columns 
print(df[['data_value', 'sample_size']].describe())
```

```python
# Average 'Data_value' for each group
segmentation = df.groupby('break_out')['data_value'].mean().sort_values(ascending=True)
plt.figure(figsize=(10, 8))
segmentation.tail(10).plot(kind='barh', color='orange')
plt.title('Segmentation: Top 10 High-Risk Groups', fontsize=14)
plt.xlabel('Average Health Risk Value (%)', fontsize=12)
plt.ylabel('Group of People', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group the data by 'Topic' (Health issues)
# 'aggregate' the Data_value by calculating the Mean
aggregation_analysis = df.groupby('topic')['data_value'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
aggregation_analysis.head(10).plot(kind='bar', color='blue')
plt.title('Aggregation Analysis: Average Risk by Health Topic', fontsize=14)
plt.xlabel('Health Topic', fontsize=12)
plt.ylabel('Aggregated Risk Value (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Aggregating by Year 
yearly_risk = df.groupby('year')['data_value'].mean()
plt.figure(figsize=(10, 5))
plt.plot(yearly_risk.index, yearly_risk.values, marker='o', color='green', linewidth=2)
plt.title('Aggregation Analysis: Risk Trend Over Time', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Risk Value (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Correlation 
cols = ['data_value', 'sample_size', 'confidence_limit_low', 'confidence_limit_high']
correlation_table = df[cols].corr()
print("--- Correlation Table ---")
print(correlation_table)
```
