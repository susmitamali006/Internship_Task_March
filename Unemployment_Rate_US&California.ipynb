```python
# There are total rows are 1559 but from those data present in fist 1534, so I tok only first 1534
df = df[0:1534]
df.shape
```

```python
# Column names 
df.columns
```

```python
# Standardise column names 
df.columns = (
    df.columns 
    .str.lower()
    .str.strip()
    .str.replace(" ","_")
)
df.columns
```

```python
# null value count
print(df.isnull().sum())
```

```python
# Duplicate count
print(df.duplicated().sum())
```

```python
# Standardise Date column
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# only filter rows where we know the Area Name is California or US
ca_data = df[df['area_name'] == 'California'].copy()
us_data = df[df['area_name'] == 'U.S.'].copy()
```

```python
# Trend Analysis for California
plt.plot(ca_data['date'], ca_data['seasonally_adjusted'], label='California',color='blue')

plt.title('Unemployment Trend Analysis California')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Year')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print(us_data.shape)
df['area_name'].value_counts()
```

```python
# Trend Analysis for US
plt.plot(us_data['date'], us_data['seasonally_adjusted'],
         label='US National', color='red')

plt.title('Unemployment Trend Analysis US')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Year')   # FIXED: Changed from plt.xlab to plt.xlabel
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()   # Better than plt.autoscale() for preventing label clipping
plt.show()
```

```python
# Comparison of California and US in single plot
plt.plot(ca_data['date'], ca_data['seasonally_adjusted'], label='California', color='blue', linewidth=2)

plt.plot(us_data['date'], us_data['seasonally_adjusted'], label='US National',color='red')

plt.title('Unemployment Trend Analysis')
plt.ylabel('Unemployment Rate (%)')
plt.xlabel('Year')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# average unemployment rate by state
state_avg = df.groupby('area_name')['seasonally_adjusted'].mean().sort_values(ascending=False)

# Create Bar Chart
plt.figure(figsize=(15, 7))   # Wide figure to fit all states
plt.bar(state_avg.index, state_avg.values, color='skyblue', edgecolor='navy')
plt.xticks(rotation=90)
plt.title('Average Unemployment Rate by State', fontsize=14)
plt.xlabel('State / Area', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
